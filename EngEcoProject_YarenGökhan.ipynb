```python
import yfinance as yf
import pandas as pd
from datetime import date

start_date = '2022-02-03'
end_date = date.today().strftime("%Y-%m-%d")

# Bist 100 ticker list  array
top10Ticker = ['BIENY.IS', 'KCHOL.IS', 'IPEKE.IS', 'AKBNK.IS', 'MGROS.IS', 'ECZYT.IS', 'SKBNK.IS', 'TUPRS.IS', 'ENKAI.IS ', 'ZOREN.IS']
ticker = top10Ticker[0]
#Storing data into data frame
df = yf.download(ticker, start_date, end_date)
df
```

```python
import plotly.graph_objects as go


fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])
fig.update_layout(
        xaxis_title="Date",
        yaxis_title="Price",
        template="plotly_white",
        title_text="Candlestick Chart"
    )

fig.show()
```

```python
import pandas_ta as ta
import numpy as np



#Calculating Bollinger Bands
df.ta.bbands(append=True)

#Calculating MFI(Money Flow Index)
df.ta.mfi(append=True)

# #Calculating RSI
df.ta.rsi(append=True)

#Calculating EMA and SMA Technical Indicators
emas_and_smas_lengths = [5,8,9,13,50,200]
for length in emas_and_smas_lengths:
  df.ta.ema(length=length, append=True)
  df.ta.sma(length=length, append=True)


# #Calculating RSI
df.ta.rsi(append=True)

#Calculating MFI(Money Flow Index)
df.ta.mfi(append=True)

#Calculating MACD
df.ta.macd(append=True)

#Calculating Ichimoku
df.ta.ichimoku(append=True)

#Calculating ADX
df.ta.adx(append=True)

# Measures the relationship between price and average price over a specific time.
df.ta.cci(append=True)

# #Calculating ATR
df.ta.atr(append=True)

standard_deviation_time = 30 #Standard Deviation for a month

df.ta.stdev(length=standard_deviation_time, append=True)

df.ta.macd(fast=34, slow=144, signal=9, append=True)
df
```

```python
#macd
df_features = pd.DataFrame()
df_features["MACD_12_26_9"] = df ["MACD_12_26_9"].to_frame().copy()
df_features["MACDh_12_26_9"] = df ["MACDh_12_26_9"].to_frame().copy()
df_features["MACDs_12_26_9"] = df ["MACDs_12_26_9"].to_frame().copy()
df_features['MACD_Signal'] = (df['MACD_12_26_9'] > df['MACDs_12_26_9']).astype(int)

#macd with 34,144,9
df.ta.macd(fast=34, slow=144, signal=9, append=True)
df_features['MACD_34_144_9'] = df['MACD_34_144_9'].to_frame().copy()
df_features['MACDh_34_144_9'] = df['MACDh_34_144_9'].to_frame().copy()
df_features['MACDs_34_144_9'] = df['MACDs_34_144_9'].to_frame().copy()

df_features
```

```python
df_features['BBL_5_2.0'] = df['BBL_5_2.0'].to_frame().copy()
df_features['BBM_5_2.0'] = df['BBM_5_2.0'].to_frame().copy()
df_features['BBU_5_2.0'] = df['BBL_5_2.0'].to_frame().copy()
df_features['BBB_5_2.0'] = df['BBB_5_2.0'].to_frame().copy()
df_features
```

```python
df_features["ADX_14"]=df["ADX_14"].to_frame().copy()
df_features
```

```python

df_features["RSI_30"] = np.where(df["RSI_14"] < 30,1,0)
df_features["RSI_70"] = np.where(df["RSI_14"] > 70,1,0)

df_features["Close"] = df["Close"].to_frame().copy()

df_features["EMA_5"] = df["EMA_5"].to_frame().copy()

df_features["SMA_9_13"] = np.where(df["SMA_9"] > df["SMA_13"],1,0)
df_features["Bollinger_Upper_Band_%"]=(df['BBU_5_2.0'] - df['Close']) / df['Close']
df_features["Signal_MFI_50"] = np.where(df["MFI_14"] < 50 ,1,0)

df_features
```

```python
df_features['target'] = np.where(df['Open'].shift(-1) - df['Close']>0 ,1,0)
df_features
```

```python
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from sklearn.preprocessing import MinMaxScaler
import optuna
from sklearn.model_selection import train_test_split


df_features = df_features.dropna()
X = df_features.drop(['target'],axis=1).values
y = df_features.target.values
train_size = int(len(X)*0.71)
X_train, X_test = X[0:train_size], X[train_size:len(X)]
y_train, y_test = y[0:train_size], y[train_size:len(X)]
```

```python
scalar=MinMaxScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.fit_transform(X_test)
```

```python
features_tomorrow = df_features.loc[df_features.index[-1], :]
features_tomorrow = features_tomorrow.drop('target')
features_tomorrow = features_tomorrow.values.reshape(1, -1)
features_tomorrow
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
def xgboost_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 50, 200),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 5),
        'random_state': 42,
    }

    model = XGBClassifier(**params)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    return accuracy

xgboost_study = optuna.create_study(direction='maximize')
xgboost_study.optimize(xgboost_objective, n_trials=20)

best_xgboost_params = xgboost_study.best_params

best_model = XGBClassifier(**best_xgboost_params)

best_model.fit(X_train, y_train)
y_pred_best_xgboost = best_model.predict(X_test)

accuracy_xgboost = accuracy_score(y_test, y_pred_best_xgboost)
precision_xgboost = precision_score(y_test, y_pred_best_xgboost, average='binary')
recall_xgboost = recall_score(y_test, y_pred_best_xgboost, average='binary')
f1_xgboost = f1_score(y_test, y_pred_best_xgboost, average='weighted')


tomorrow_pred = best_model.predict(features_tomorrow)
print( f'Tomorrow prediction :{tomorrow_pred}(1->will increase 0->will decrease)')
print(f' XGBoost model accuracy is: {accuracy_xgboost}')
print(f' XGBoost model precision is: {precision_xgboost}')
print(f' XGBoost model recall is: {recall_xgboost}')
print(f' XGBoost model f1 measure is: {f1_xgboost}')
```

```python


def objective(trial):
    # Define the hyperparameters to be optimized and their search spaces
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 3, 10)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 5)

    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)

    return accuracy

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20)

best_params = study.best_params

best_model = RandomForestClassifier(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    min_samples_leaf=best_params['min_samples_leaf'],
    random_state=42
)

best_model.fit(X_train, y_train)

y_pred_best = best_model.predict(X_test)


accuracy = accuracy_score(y_test, y_pred_best)
precision = precision_score(y_test, y_pred_best, average='binary')
recall = recall_score(y_test, y_pred_best, average='binary')
f1 = f1_score(y_test, y_pred_best, average='weighted')


tomorrow_pred = best_model.predict(features_tomorrow)
print( f'Tomorrow prediction :{tomorrow_pred}(1->will increase 0->will decrease)')
print(f' RandomForest model accuracy is :{accuracy}' )
print(f' RandomForest model precision is :{precision}')
print(f' RandomForest model recall is :{recall}')
print(f' RandomForest model f1measure is :{f1}')

print(f"Predictions:{y_pred_best}")
print(f"Real results:{y_test}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential()
model.add(LSTM(100, activation="tanh", return_sequences=True, input_shape=(1, X_train.shape[1])))
model.add(LSTM(49, activation="tanh"))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer="rmsprop", loss='binary_crossentropy', metrics=['accuracy'])

X_train = np.resize(X_train, (X_train.shape[0],1,X_train.shape[1]))
X_test = np.resize(X_test, (X_test.shape[0],1,X_test.shape[1]))
model.fit(X_train, y_train, epochs=10, batch_size=len(X_train), validation_data=(X_test, y_test), verbose=0)
threshold = 0.525
y_pred = model.predict(X_test).T[0]
print(y_pred)
y_pred = (y_pred > threshold).astype("int32")

accuracy = accuracy_score(y_test, y_pred)
print(f'model accuracy is :{accuracy} ' )

precision = precision_score(y_test, y_pred, average='binary')
print(f'model precision is :{precision}')

recall = recall_score(y_test, y_pred, average='binary')
print(f'model recall is :{recall}')

f1 = f1_score(y_test, y_pred, average='weighted')
print(f'model f1measure is :{f1}')

tomorrow_pred = best_model.predict(features_tomorrow)
print( f'Tomorrow prediction :{tomorrow_pred}(1->will increase 0->will decrease)')



```
